```python
temperatures = [25, 28, 30, 22, 35, 40, 18, 24]

# Konversi ke Fahrenheit
temp_fahrenheit = [(C * 9/5) + 32 for C in temperatures]

# Ambil 3 data terakhir
last_3 = temp_fahrenheit[-3:]

print("Fahrenheit:", temp_fahrenheit)
print("3 data terakhir:", last_3)
```
